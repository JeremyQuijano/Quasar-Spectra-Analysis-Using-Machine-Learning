```python
# Import necessary packages
from astropy.io import fits
import tensorflow as tf
import numpy as np
import cv2
import glob
import random
```

```python
# Create path to files
CATEGORIES = ["spectra_0.55HM12_zr9_z5.000_1D", "spectra_1.0HM12_zr9_z5.000_1D", "spectra_1.8HM12_zr9_z5.000_1D"]
DIR = ['/Users/quijeremy/desktop/senior_thesis/spectra_data/test_data/spectra_0.55HM12_zr9_z5.000_test/*.fits',
      '/Users/quijeremy/desktop/senior_thesis/spectra_data/test_data/spectra_1.0HM12_zr9_z5.000_test/*.fits',
      '/Users/quijeremy/desktop/senior_thesis/spectra_data/test_data/spectra_1.8HM12_zr9_z5.000_test/*.fits']

def prepare(filepath):
    #noise = np.random.normal(0, 0.2, 2048)
    #scale = 0.8
    img_array = fits.getdata(filepath)  # read in the image, convert to grayscale
    return img_array.reshape(-1, 2048, 1)  # return the image with shaping that TF wants.
```

```python
# Load the models
model1  = tf.keras.models.load_model("3-conv-16-nodes-0.00noise-1559002311")
model2  = tf.keras.models.load_model("3-conv-16-nodes-0.01noise-1559010347")
model3  = tf.keras.models.load_model("3-conv-16-nodes-0.02noise-1559012543")
model4  = tf.keras.models.load_model("3-conv-16-nodes-0.05noise-1559018671")
model5  = tf.keras.models.load_model("3-conv-16-nodes-0.10noise-1559019795")
model6  = tf.keras.models.load_model("3-conv-16-nodes-0.15noise-1559674545")
model7  = tf.keras.models.load_model("3-conv-16-nodes-0.20noise-1559022405")
```

```python
# Make sure all data was imported
spectra1_test = glob.glob(DIR[0])
print(len(spectra1_test))

spectra2_test = glob.glob(DIR[1])
print(len(spectra2_test))

spectra3_test = glob.glob(DIR[2])
print(len(spectra3_test))
```

```python
# Test with data that was separate from training set
for image in spectra1_test:
    prediction = model1.predict([prepare(image)])
    #print(prediction)  # will be a list in a list.
    print(CATEGORIES[int(np.argmax(prediction[0]))])
```
