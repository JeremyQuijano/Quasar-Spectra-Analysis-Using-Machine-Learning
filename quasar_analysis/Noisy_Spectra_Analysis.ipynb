```python
# Necessary Packages
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import glob
import random
from astropy.io import fits
```

```python
# Import the necessary file
spectra1 = glob.glob('/Users/quijeremy/desktop/senior_thesis/spectra_data/full_spectra_files/spectra_0.55HM12_zr9_z5.000.fits')
spectra2 = glob.glob('/Users/quijeremy/desktop/senior_thesis/spectra_data/full_spectra_files/spectra_1.0HM12_zr9_z5.000.fits')
spectra3 = glob.glob('/Users/quijeremy/desktop/senior_thesis/spectra_data/full_spectra_files/spectra_1.8HM12_zr9_z5.000.fits')
```

```python
# Make sure file paths are correct
print(spectra1)
print(spectra2)
print(spectra3)
```

```python
# Get the imgae data from fits file
spectra1_data = np.array([fits.getdata(image) for image in spectra1])
spectra2_data = np.array([fits.getdata(image) for image in spectra2])
spectra3_data = np.array([fits.getdata(image) for image in spectra3])
```

```python
# Create noisy data
spectra1_noisy_01 = []
spectra1_noisy_02 = []
spectra1_noisy_05 = []
spectra1_noisy_10 = []
spectra1_noisy_15 = []
spectra1_noisy_20 = []

for image in spectra1:
    noise_01 = np.random.normal(0, 0.01, 2048)
    noise_02 = np.random.normal(0, 0.02, 2048)
    noise_05 = np.random.normal(0, 0.05, 2048)
    noise_10 = np.random.normal(0, 0.1, 2048)
    noise_15 = np.random.normal(0, 0.15, 2048)
    noise_20 = np.random.normal(0, 0.2, 2048)
    spectra1_noisy_01.append(fits.getdata(image) + noise_01)
    spectra1_noisy_02.append(fits.getdata(image) + noise_02)
    spectra1_noisy_05.append(fits.getdata(image) + noise_05)
    spectra1_noisy_10.append(fits.getdata(image) + noise_10)
    spectra1_noisy_15.append(fits.getdata(image) + noise_15)
    spectra1_noisy_20.append(fits.getdata(image) + noise_20)
```

```python
# Create noisy data
spectra3_noisy_01 = []
spectra3_noisy_02 = []
spectra3_noisy_05 = []
spectra3_noisy_10 = []
spectra3_noisy_15 = []
spectra3_noisy_20 = []

for image in spectra1:
    noise_01 = np.random.normal(0, 0.01, 2048)
    noise_02 = np.random.normal(0, 0.02, 2048)
    noise_05 = np.random.normal(0, 0.05, 2048)
    noise_10 = np.random.normal(0, 0.1, 2048)
    noise_15 = np.random.normal(0, 0.15, 2048)
    noise_20 = np.random.normal(0, 0.2, 2048)
    spectra3_noisy_01.append(fits.getdata(image) + noise_01)
    spectra3_noisy_02.append(fits.getdata(image) + noise_02)
    spectra3_noisy_05.append(fits.getdata(image) + noise_05)
    spectra3_noisy_10.append(fits.getdata(image) + noise_10)
    spectra3_noisy_15.append(fits.getdata(image) + noise_15)
    spectra3_noisy_20.append(fits.getdata(image) + noise_20)
```

```python
# Plot Clean Data
randnum = random.randint(0,4999)
plt.plot(spectra1_data[0,0,:])
plt.title("spectra_0.55HM12_zr9_z5.000_clean_" + str(randnum))
plt.show()
```

```python
# Plot scaled data
plt.figure(figsize=(15,3))
plt.title('Sample Plot with Different Scale Factors', size=16)
plt.plot(spectra1_data[0,randnum,:]*0.8, label='0.8')
plt.plot(spectra1_data[0,randnum,:]*0.9, label='0.9')
plt.plot(spectra1_data[0,randnum,:]*1.0, label='1.0')
plt.plot(spectra1_data[0,randnum,:]*1.1, label='1.1')
plt.plot(spectra1_data[0,randnum,:]*1.2, label='1.2')
plt.xticks(size=12)
plt.yticks(size=12)
plt.legend(loc='upper left')
plt.savefig('sample_scale.png')
plt.show()
```

```python
# Plot noisy data
fig = plt.figure(figsize=(15, 5))
plt.xticks([])
plt.yticks([])
plt.title('Increasing Noise in Data', size=16)
ax = fig.subplots(2,3)

ax[0,0].plot(spectra1_noisy_01[0][100][:])
ax[0,1].plot(spectra1_noisy_02[0][100][:])
ax[0,2].plot(spectra1_noisy_05[0][100][:])
ax[1,0].plot(spectra1_noisy_10[0][100][:])
ax[1,1].plot(spectra1_noisy_15[0][100][:])
ax[1,2].plot(spectra1_noisy_20[0][100][:])
plt.savefig('sample_noise.png')
plt.show()
```
