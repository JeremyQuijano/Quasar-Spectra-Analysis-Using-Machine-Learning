```python
# Import necessary packages
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create Accuracy vs Noise Plot with same nosie scale
"""
We should see the accuracy decrease as the noise of the data increases
- Measuring the accuracy of the model after 15 epochs
- Taking both the training and validation accuracy from the 15th epoch
"""
noise               = [0, .01, .02, .05, .10, .15, .20]
training_acc        = [.9606, .9099, .8931, .8156, .7550, .7450, .3291]
valid_acc           = [.9496, .8919, .8311, .7678, .6778, .6274, .3252]
test_same_noise_acc = [.9607, .8993, .8453, .7633, .7233, .6413, .3333]

plt.figure(figsize=(13,5))
plt.title('Accuracy with Same Noise Input', size=24)
plt.xticks([0, .01, .02, .05, .1, .15, .2], size=12)
plt.plot(noise, training_acc, label='training', marker='o', linestyle='--')
plt.plot(noise, valid_acc, label='validation', marker='o', linestyle='--')
plt.plot(noise, test_same_noise_acc, label='test', marker='o', linestyle='--')
plt.xlabel('Noise', size=24)
plt.ylabel('Accuracy', size=24)
plt.legend(loc='upper right', fontsize='x-large')
plt.savefig('accracy_vs_same_noise.png')
plt.show()
```

```python
# Create Model Accuracy vs Noise plot for differing noise scales
noise           = [    0,   .01,   .02,   .05,   .10,   .15,   .20]
test_model1_acc = [.9607, .3333, .3333, .3333, .3333, .3333, .3333]
test_model2_acc = [.8800, .9020, .7567, .3753, .3340, .3333, .3333]
test_model3_acc = [.8487, .8500, .8453, .5800, .3600, .3373, .3333]
test_model4_acc = [.6987, .7020, .7100, .7633, .5540, .3600, .3587]
test_model5_acc = [.5773, .5780, .5853, .6447, .7233, .5787, .4173]
test_model6_acc = [.6327, .6300, .6347, .6347, .6480, .6413, .5953]
test_model7_acc = [.3333, .3333, .3333, .3333, .3333, .3333, .3333]

# Plot it!
plt.figure(figsize=(15,5))
plt.title('Test Accuracy with Different Noise Input', size=24)
plt.xticks([0, .01, .02, .05, .1, .15, .2], size=12)
plt.plot(noise, test_model1_acc, label='Trained on 0.00 noise', marker='o', linestyle='--')
plt.plot(noise, test_model2_acc, label='Trained on 0.01 noise', marker='o', linestyle='--')
plt.plot(noise, test_model3_acc, label='Trained on 0.02 noise', marker='o', linestyle='--')
plt.plot(noise, test_model4_acc, label='Trained on 0.05 noise', marker='o', linestyle='--')
plt.plot(noise, test_model5_acc, label='Trained on 0.10 noise', marker='o', linestyle='--')
plt.plot(noise, test_model6_acc, label='Trained on 0.15 noise', marker='o', linestyle='--')
plt.plot(noise, test_model7_acc, label='Trained on 0.20 noise', marker='o', linestyle='--')
plt.xlabel('Noise', size=24)
plt.ylabel('Accuracy', size=24)
plt.legend(loc='upper right', fontsize='large')
plt.savefig('accracy_vs_different_noise.png')
plt.show()
```

```python
# Create Accuracy vs Multiple plot for different scaling factors
scale                  = [  0.8,   0.9,   1.0,   1.1,   1.2]
test_model1_scale_acc  = [.9207, .9533, .9607, .9620, .9600]

# Plot it!
plt.figure(figsize=(7, 4))
plt.title('Accuracy with Different Scaling Factor', size=14)
plt.xticks([0.8, 0.9, 1.0, 1.1, 1.2])
plt.plot(scale, test_model1_scale_acc, label='Trained on 1.0 Scale', marker='o', linestyle='--')
plt.xlabel('Scale Factor', size=14)
plt.ylabel('Accuracy', size=14)
plt.savefig('accracy_vs_scaling.png')
plt.show()
```

```python
# Create skew bias plot for same noise scales
skew0_55_noise   = [14, 28,  35,  30,  60,  85, 500]
skew1_8_noise    = [16, 63, 154, 174, 142, 212,   0]
model_acc_noise  = [.9400, .8180, .6220, .5920, .5960, .4060, 0]

index = np.arange(7)
bar_width = 0.35

plt.figure(figsize=(10,3))

rects1 = plt.bar(index, skew0_55_noise, bar_width,
color='blue',
label='0.55HM12_zr9')

rects2 = plt.bar(index + bar_width, skew1_8_noise, bar_width,
color='orange',
label='1.8HM12_zr9')

plt.xlabel('Noise', size=16)
plt.ylabel('Skew', size=16)
plt.title('Test Skew Bias with Same Noise', size=16)
plt.xticks(index + bar_width, ('0.00', '0.01', '0.02', '0.05', '0.10', '0.15', '0.20'))
plt.legend(loc='upper left', fontsize='large')

plt.tight_layout()
plt.savefig('skew_bias_same_noise.png')
plt.show()

plt.figure(figsize=(10, 5))
plt.title('Model Accuracy with 1.0 Spectra', size=14)
plt.xticks([0.00, 0.01, 0.02, 0.05, 0.10, 0.15, 0.20])
plt.plot(noise, model_acc_noise, marker='o', linestyle='--')
plt.xlabel('Noise', size=14)
plt.ylabel('Accuracy', size=14)
plt.savefig('skew_acc_noise.png')
plt.show()
```

```python
# Create skew bias plot for different scale factors
skew0_55_scale  = [  2,   7,  14,  22,  26]
skew1_8_scale   = [ 48,  20,  16,  10,   4]
model_acc_scale = [.9000, .9460, .9400, .9360, .9400]
scale = [0.8, 0.9, 1.0, 1.1, 1.2]

index = np.arange(5)
bar_width = 0.35

plt.figure(figsize=(10, 3))

rects1 = plt.bar(index, skew0_55_scale, bar_width,
color='blue',
label='0.55HM12_zr9')

rects2 = plt.bar(index + bar_width, skew1_8_scale, bar_width,
color='orange',
label='1.8HM12_zr9')

plt.xlabel('Scale Factor', size=16)
plt.ylabel('Skew', size=16)
plt.title('Test Skew Bias with Different Scale Factors', size=16)
plt.xticks(index + bar_width, ('0.8', '0.9', '1.0', '1.1', '1.2'))
plt.legend(loc='upper right', fontsize='large')
plt.tight_layout()
plt.savefig('skew_bias_diff_scale.png')
plt.show()

plt.figure(figsize=(10, 5))
plt.title('Model Accuracy with 1.0 Spectra', size=14)
plt.xticks([0.8, 0.9, 1.0, 1.1, 1.2])
plt.plot(scale, model_acc_scale, marker='o', linestyle='--')
plt.xlabel('Scale Factor', size=14)
plt.ylabel('Accuracy', size=14)
plt.savefig('skew_acc_scale.png')
plt.show()
```
